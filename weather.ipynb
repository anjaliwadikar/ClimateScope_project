```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/weather.csv'

df = pd.read_csv(file_path)
print(df.head())
```

```python

# Step 3: Inspect Data Structure, Data Types, and Key Variables
print(" Dataset Shape:", df.shape)
print("\n Data Info:")
df.info()
print("\n Summary Statistics (numeric columns):")
display(df.describe())
print("\n Data Types:")
print(df.dtypes)
```

```python
# Step 4: Missing Values, Anomalies & Data Coverage (short)
print(" Missing Values:", "None " if df.isnull().sum().sum()==0 else df.isnull().sum()[df.isnull().sum()>0])
print("Humidity anomalies:", len(df.query("humidity<0 or humidity>100")),
      "| Temperature anomalies:", len(df.query("temperature_celsius<-90 or temperature_celsius>60")))
print("\nData coverage:\n", df.count())
```

```python

print(" Missing values before cleaning:")
print(df.isna().sum()[df.isna().sum() > 0])
# Handling missing/inconsistent data
df.loc[(df["temperature_celsius"] < -50) | (df["temperature_celsius"] > 60), "temperature_celsius"] = None
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna("Unknown", inplace=True)
print("\n Missing values after cleaning:")
print(df.isna().sum()[df.isna().sum() > 0])
print("\nCleaned Data (first 5 rows):")
print(df.head())
```

```python

df["last_updated"] = pd.to_datetime(df["last_updated"], errors="coerce")
# Daily data
daily_data = df.copy()
print(daily_data.head())
```

```python
# Add year-month column
df["year_month"] = df["last_updated"].dt.to_period("M").astype(str)

# Monthly averages per country
monthly_data = (
    df.groupby(["country", "year_month"])
      .mean(numeric_only=True)
      .reset_index()
)

print(monthly_data.head())
```

```python
# Import libraries
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Statistical Analysis
print("Summary statistics for key variables:")
print(df[['temperature_celsius','humidity','precip_mm','wind_kph','gust_kph']].describe())

# Plot histograms for each key variable
variables = ['temperature_celsius','humidity','precip_mm','wind_kph','gust_kph']

for var in variables:
    fig = px.histogram(df, x=var, nbins=50,
                       title=f'Distribution of {var.replace("_", " ").title()}',
                       labels={var: var.replace("_", " ").title()})
    fig.update_layout(template='plotly_white',
                      title_font_size=22,
                      xaxis_title_font_size=16,
                      yaxis_title_font_size=16,
                      hovermode='x unified')
    fig.show()
```

```python
# Correlation matrix
numeric_cols = ['temperature_celsius','humidity','precip_mm','wind_kph','gust_kph']
corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Weather Variables")
plt.show()
```

```python
#  Seasonal Patterns (Temperature Trends)

df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month

monthly_avg = df.groupby(['year','month'])['temperature_celsius'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=monthly_avg, x='month', y='temperature_celsius', hue='year', palette='tab10')
plt.title("Seasonal Temperature Trends by Year")
plt.ylabel("Avg Temperature (Â°C)")
plt.show()
```

```python
# Group by country and calculate the average wind speed
country_wind = df.groupby('country')['wind_kph'].mean().reset_index()

# Sort in descending order and get the top 10 countries
top_10_windy_countries = country_wind.sort_values(by='wind_kph', ascending=False).head(10)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_windy_countries, y='country', x='wind_kph', palette='viridis', hue='country', legend=False) # Added hue='country' and legend=False
plt.title('Top 10 Countries with Highest Average Wind Speed')
plt.xlabel('Average Wind Speed (kph)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#  Regional Comparison (Avg Temperature by Country)
country_temp = df.groupby('country')['temperature_celsius'].mean().reset_index()

fig = px.choropleth(
    country_temp,
    locations="country",
    locationmode="country names",
    color="temperature_celsius",
    color_continuous_scale="RdYlBu_r",
    title="Average Temperature by Country"
)
fig.show()
```

```python
# Scatter Plot: Temperature vs. Feels Like Temperature

plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature_celsius', y='feels_like_celsius', data=df)
plt.title('Relationship Between Actual and Perceived Temperature')
plt.xlabel('Actual Temperature (Â°C)')
plt.ylabel('Feels Like Temperature (Â°C)')
plt.tight_layout()
plt.show()
```

```python
#  Humidity Distribution by Cloud Cover

plt.figure(figsize=(15, 8))
sns.lineplot(
data=df,
x='cloud', y='humidity', errorbar=None, marker='o' )      # update- replaces ci=None
plt.title('Humidity Trend by Cloud Cover')
plt.xlabel('Cloud Cover (%)')
plt.ylabel('Humidity (%)')
plt.tight_layout()
plt.show()
```

```python
# Linegraph: Monthly Variation
monthly_matrix = df.groupby(['month','year'])['temperature_celsius'].mean().unstack()

plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_matrix)
plt.title("Monthly Temperature Variation (Â°C)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
# Stacked Bar Chart: Weather Conditions by Month

# First, prepare the data by counting conditions per month
df['month'] = df['last_updated'].dt.month
condition_counts = df.groupby(['month', 'condition_text']).size().reset_index(name='count')
pivot_df = condition_counts.pivot(index='month', columns='condition_text', values='count').fillna(0)

pivot_df.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Weather Conditions by Month')
plt.xlabel('Month')
plt.ylabel('Count of Conditions')
plt.xticks(rotation=0)
plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
